### Weather scraping
##### As the weather information is in a table format on the website, I can use pd.html function to read the table format and extract into a list

```python
df_list=[]
for i in range(1,13):
    url = 'https://au.freemeteo.com/weather/washington/history/monthly-history/?gid=4140963&station=19064&month='+str(i)+'&year=2018&language=english&country=us-united-states'
    print(url)
    df1=pd.read_html(url)
    df_list.append(df1[6].iloc[:,:10])
```

```python
#converting the list into a dataframe
df = pd.DataFrame(df_list)
```

```python
df_large = pd.concat(df_list, ignore_index=True)  
df_large
```

```python
df_large.columns
```

```python
#droping unwanted columns
df_large.drop(['Maximum steady wind', 'Maximum wind gust', 'Total daily precipitation',
       'Snow depth', 'Pressure', 'Icon'], axis=1, inplace=True)
```

```python
#dataframe with 2018 weather data only
df_large.head(2)
```

```python
#converting date column into datetime format
df_large.Date = pd.to_datetime(df_large.Date)
```

```python
#removing characters from columns
df_large['Daily minimum temperature']=df_large['Daily minimum temperature'].str.strip('°C	')

df_large['Daily maximum temperature']=df_large['Daily maximum temperature'].str.strip('°C	')
```

```python
#converting into float columns
df_large['Daily minimum temperature']=df_large['Daily minimum temperature'].map(lambda x: float(x) if x is not np.nan else x)

df_large['Daily maximum temperature']=df_large['Daily maximum temperature'].map(lambda x: float(x) if x is not np.nan else x)
```

```python
df_large['Description'] = df_large.Description.replace('weather events not reported','Clear')
#df_large['Description'] = weather_df.Description.replace('0.0','Clear')
```

```python
#new column with average temperature
df_large['avg_temp'] = df_large.mean(axis=1)
```

### Concatenating all the dataframes together

```python
#combining all 2018 bike share data into a single data frame
df_2018=pd.concat(to_add, axis=0, ignore_index=True)
```

```python
#converting date column into datetime format to extract month, hour and data columns
df_2018['Start date']= pd.to_datetime(df_2018['Start date'])
```

```python
df_2018['date']=df_2018['Start date'].dt.date

df_2018['time'] = df_2018['Start date'].dt.time

df_2018['hour'] =df_2018['Start date'].dt.hour
```

```python
df_2018.head()
```

```python
df_2018 = df_2018[['Duration','Member type','date','hour']]
```

```python
df_2018.head(2)
```

```python
#New dataframe of total number of bike uses per day and hour
grouped_df = pd.DataFrame(df_2018.groupby([ "date", "hour",'Member type']).agg({'Duration':'sum',
                                                                 'Duration':'count'
                                                                 })).reset_index()
```

```python
grouped_df.shape
```

```python
#Converting membertype into binary
member_df = pd.get_dummies(grouped_df['Member type'], drop_first=True)
# member = 1
# casual membership = 0
```

```python
grouped_df = pd.concat([grouped_df, member_df], axis=1)
```

```python
grouped_df = grouped_df.drop('Member type', axis=1)
```

```python
#Dataframe with relevant information about 2018 bikeshare data
grouped_df.to_csv('hour_grouped.csv', index=False)
```
